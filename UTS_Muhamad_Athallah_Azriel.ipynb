```python
oi = az
```

```python
oi = az.iloc[:,7: 82]
```

```python
from sklearn.model_selection import train_test_split
```

```python
oi_train, oi_test, at_train, at_test = train_test_split(oi, at, test_size = 0.20, random_state = 82)
```

```python
from sklearn.tree import DecisionTreeClassifier as arn
```

```python
dtree = arn()
```

```python
dtree.fit(oi_train, at_train)
```

```python
at_pred = dtree.predict(oi_test)
```

```python
at_pred
```

```python
from sklearn.metrics import accuracy_score as ps
```

```python
ps (at_test, at_pred)
```
